# Example 4: LASSO

LASSO (Least Absolute Shrinkage and Selection Operator) is a regression analysis method that performs both variable selection and regularization in order to enhance the prediction accuracy and interpretability. 

LASSO works by adding a penalty term to the loss function, which is the absolute value of the coefficients multiplied by a $\lambda$ parameter. This penalty term encourages the model to not only fit the data well but also to keep the model coefficients as small as possible, just like ridge regression. The difference between ridge regression and LASSO, however, comes about from the use of the absolute value, which can lead to some coefficients being exactly zero. This makes LASSO particularly useful in scenarios where the number of predictors is very high compared to the number of observations or when a model with fewer variables is preferred for interpretation or implementation reasons.

```python
# Standard imports
import numpy as np
import matplotlib.pyplot as plt

# Needed to perform ordinary linear regression and ridge regression
from sklearn.linear_model import LinearRegression, Lasso

# Metrics used to evaluate a model
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

Below, we generate 100 data points with five predictors, $X_1, \ldots, X_5$, and a dependent variable whose true model is $$Y = 2X_1 + 3X_2 + X_5 + \epsilon.$$

```python
# Generate some sample data
np.random.seed(4)
features = 5
X = np.random.rand(100, 5)  # 100 samples with 5 features

# Ground Truth
Y = 2 * X[:, 0] + 3 * X[:, 2] + X[:, 4] + np.random.randn(100)
```

As with ridge regression, we will split our dataset into a training, validation, and testing set for the same reasons. In fact, you will see this method in almost every modeling technique covered in the class. 

```python
# Split the data into training, validation, and testing sets
X_train, X_temp, Y_train, Y_temp = train_test_split(X, Y, test_size=0.3, 
                                                    random_state=4)
X_val, X_test, Y_val, Y_test = train_test_split(X_temp, Y_temp, test_size=0.5, 
                                                random_state=4)

# For fitting the function once hyperparameters are selected
X_train_val = np.concatenate((X_train, X_val))
Y_train_val = np.concatenate((Y_train, Y_val))
```

We now fit the data to a LASSO model with $\lambda = 0.05$. Notice that the syntax of the coding is nearly identical to what we've seen with both linear regression and ridge regression. This is because of the standardization of modeling done in the Scitkit-Learn library. 

Notice that our regularization parameter is named `alpha`. Again, $\alpha = \lambda$, so these parameter variables are interchangeable. 

```python
# Set the regularization parameter lambda (alpha)
lasso = Lasso(alpha=0.05)  
lasso.fit(X_train, Y_train)

# Print Coefficients
lasso_coefs = lasso.coef_
print("LASSO Coefficients:", lasso_coefs)
print("LASSO Intercept:", lasso.intercept_)
```

The mode we get with LASSO, setting $\lambda = 0.05$ is $$\hat Y_{\lambda = 0.05} = 1.64 + 1.38X_1 + 1.46X_3 + 0.20X_5.$$

Now, lets attempt to find the optimal value for our parameter $\lambda$. Our method is the same as with ridge regression. 

Here, we will use the mean absolute error instead of the mean squared error. In truth, there are several possible metrics one could use when selecting a parameter--each with its own strengths and weaknesses. We will cover their differences at a later point in the course. For now, know that the mean absolute error is calculated as: 

$$MAE = \frac{1}{n} \sum_{i = 1}^n \left| \hat y - y \right|$$

or the absolute value of the predicted versus actual values of $y$, sumed and divided by the number of samples in the set. Below, the set we are considering is the validation set. 

```python
# Create a range of lambda values
lambdas = np.arange(0.01, 0.31, 0.01)

# Initialize arrays to store coefficients and MSE values
coefficients = []
mae_values = []

# Fit LASSO models and calculate MSE for different lambda values
for lam in lambdas:
    lasso = Lasso(alpha=lam)
    lasso.fit(X_train, Y_train) # fit with training set
    coefficients.append(lasso.coef_)
    
    Y_pred = lasso.predict(X_val)
    mae = mean_absolute_error(Y_val, Y_pred) #evaluate with validation set
    mae_values.append(mae)

# Convert coefficients and MSE values to numpy arrays for easier manipulation
coefficients = np.array(coefficients)
mae_values = np.array(mae_values)
```

We now find the optimal $\lambda$ based on the smallest mean absolute error. 

```python
# Lambda value corresponding to the model with lowest validation MSE
best_lambda = round(lambdas[np.argmin(mae_values)], 3)
print(best_lambda)
```

As before, we graph the coefficients as well as the error. Notice that the trajectory of the coefficients is different than ridge regression--the coefficients quickly go to zero--and this is because of the absolute value in using in the loss function for LASSO.

```python
# Create a subplot with two plots (coefficient paths and MSE values)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), sharex=True)

# Plot the coefficient paths
ax1.axvline(x=best_lambda, color='red', linestyle='--', 
            label=f'Best Lambda ({best_lambda})')
for feature in range(features):
    ax1.plot(lambdas, coefficients[:, feature], 
             label=f'Feature {feature + 1}')
ax1.set_ylabel('Coefficient Magnitude')
ax1.set_title('LASSO Coefficient Paths')
ax1.legend()
ax1.grid()

# Plot the MAE values
ax2.plot(lambdas, mae_values, color='blue', label='MAE')
ax2.axvline(x=best_lambda, color='r', linestyle='--', 
            label=f'Best Lambda ({best_lambda})')
ax2.set_xlabel('lambda')
ax2.set_ylabel('Mean Absolute Error (MAE)')
ax2.set_title(f'Mean Absolute Error (MAE) vs. LASSO ({best_lambda})')
ax2.legend()
ax2.grid()

plt.tight_layout()
plt.show()
```

We now compute the final model. 

```python
# Compute the final model
lasso = Lasso(alpha=best_lambda) 
lasso.fit(X_train_val, Y_train_val)

print("LASSO Coefficients:", lasso.coef_)
print("LASSO Intercept:", lasso.intercept_)
```

The model we calculated is 
$$\hat Y = 1.06 + 1.52X_1 + 1.91X_3 - 0.27X_4 + 0.86X_5.$$

Finally, we evaluate our model as well as the ordinary least squares linear regression model for comparison. One of the interesting observations with this example is that OLS has *worse* metrics than our LASSO model. Although OLS Linear Regression is BLUE--the best linear unbiased estimator--of the underlying model, it *can* be beat by *biased* estimators. LASSO is a biased estimator that achieves this. 

```python
# Evaluate the final model
Y_test_pred = lasso.predict(X_test)
mae = mean_absolute_error(Y_test, Y_test_pred)
r2 = r2_score(Y_test, Y_test_pred)
print(f"The testing set MAE is {mae}")
print(f"The testing set coefficient of determination (R^2) is {r2}")
```

```python
# For comparison to Ordinary Least Squares Linear Regression

# Fit model to same dataset
ols = LinearRegression()
ols.fit(X_train_val, Y_train_val)

# Find same scores
Y_test_pred = ols.predict(X_test)
mse = mean_absolute_error(Y_test, Y_test_pred)
r2 = r2_score(Y_test, Y_test_pred)
print(f"The testing set MAE for OLS Linear Regression is {mse}")
print(f"The testing set coefficient of determination (R^2) for" + 
      " OLS Linear Regression is {r2}")
```
